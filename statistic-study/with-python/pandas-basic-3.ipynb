```python
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
import statsmodels.regression as smrg
import statsmodels as sm
import matplotlib.pyplot as plt

def i(v):
    print(type(v))
    print(v)
```

```python
x:np.ndarray = np.arange(100)
# i(x)
y:np.ndarray = np.random.randn(len(x)) + 0.5 * x + 20
df:pd.DataFrame = pd.DataFrame({"x":x, "y":y})

df.plot.bar(x="x", y="y")
plt.show()
```

```python
model: smrg.linear_model.RegressionResultsWrapper = smf.ols("y~x", data=df).fit()
summary:sm.iolib.summary.Summary = model.summary()
# i(summary)
summary
```
